```python
read1 = pd.read_csv('1-STAR_Reading_Assessment_Detail (1).csv', skiprows=1)
read2 = pd.read_csv('2-STAR_Reading_Assessment_Detail.csv', skiprows=1)
read3 = pd.read_csv('3-STAR_Reading_Assessment_Detail (9).csv', skiprows=1)
read4 = pd.read_csv('4-STAR_Reading_Assessment_Detail (8).csv', skiprows=1)
read5 = pd.read_csv('5-STAR_Reading_Assessment_Detail (7).csv', skiprows=1)
read6 = pd.read_csv('6-STAR_Reading_Assessment_Detail (5).csv', skiprows=1)
```

```python
reads = [read1, read2, read3, read4, read5, read6]
```

```python
read_data = pd.concat(reads)
```

```python
read_data.describe()
```

```python
read_data.pivot_table('AssessmentGrade', 'StudentId', aggfunc=np.mean).reset_index()
```

```python
read_data
```

```python
read_data['DateTaken'].dtype
```

```python
read_data['DateTime'] = pd.DatetimeIndex(read_data['DateTaken'])
```

```python
pd.to_datetime(pd.DatetimeIndex(read_data['DateTaken'])).year
```

##Create GE - Assessment grade (difference between current grade and reading level)

```python
read_data['GE-Grade'] = read_data['GE'] - read_data['AssessmentGrade']
```

```python
clean = read_data.pivot_table('GE-Grade', 'StudentId', aggfunc=np.mean).reset_index()
```

```python
ge_max = read_data.pivot_table('GE-Grade', 'StudentId', aggfunc=np.max).reset_index()
```

```python
clean['GE-Grade_MAX'] = ge_max['GE-Grade']
```

```python
clean.head()
```

```python
# Commented out because we want to separate student info by grade level as well
# clean.to_csv(path_or_buf='reading_level.csv')
```

```python
clean['GE-Grade'].hist(bins=25, figsize=(16,8))
```

```python
clean['GE-Grade_MAX'].hist(bins=25, figsize=(16,8))
```

## Create Pivot on StudentId and school year (AssessmentGrade)

```python
average = read_data.pivot_table('GE-Grade', ['StudentId', 'AssessmentGrade'], aggfunc=np.mean).reset_index()
```

```python
ge_diff_max = read_data.pivot_table('GE-Grade', ['StudentId','AssessmentGrade'], aggfunc=np.max).reset_index()
```

```python
average['GE-Grade_MAX'] = ge_diff_max['GE-Grade']
```
